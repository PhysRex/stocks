# finviz api download

```python
from finvizfinance.screener.valuation import Valuation

foverview = Valuation()

filters_dict = {
    'Exchange':'Any',
    'Country': 'USA',
    'Market Cap.': '+Large (over $10bln)',
    'Industry': 'Stocks only (ex-Funds)',
}

foverview.set_filter(filters_dict=filters_dict)

df = foverview.screener_view()
```

```python
df.shape
```

```python
sorted_stocks = df.sort_values('Market Cap', ascending=False)
sorted_stocks.head()
```

```python
positive_stocks = sorted_stocks[sorted_stocks['P/E'] > 0]
# positive_stocks.head()
positive_stocks.shape
```

```python
highest_earning_stocks = positive_stocks[positive_stocks['P/E'] < (0.5* positive_stocks['Market Cap'] / 1000000000)]
highest_earning_stocks.head()
highest_earning_stocks.shape
```

```python
print(highest_earning_stocks.columns.tolist())
```

```python
analysis_stocks = highest_earning_stocks[['Ticker', 'Price', 'Market Cap', 'P/E', 'EPS Past 5Y', ]].copy()
analysis_stocks.head()
```

```python
analysis_stocks.loc[:, ['Earnings (in B)']] = analysis_stocks['Market Cap'] / (1000000000 * analysis_stocks['P/E'])
analysis_stocks.head(10)
```

```python
analysis_stocks = analysis_stocks[['Ticker', 'Price', 'Earnings (in B)', 'EPS Past 5Y', ]].copy()
analysis_stocks.rename(columns={'EPS Past 5Y': 'Earnings Growth'}, inplace=True)
# earnings_growth_col = analysis_stocks.pop('Earnings Growth')
# df.insert(3, earnings_growth_col.name, earnings_growth_col)  # Is in-place

analysis_stocks.head(10)
```

```python
import pandas as pd
# from pandas import read_excel

sheet = 'constituents' # change it to your sheet name, you can find your sheet name at the bottom left of your excel file
file_name = 's&p500-constituents.xlsx' # change it to the name of your excel file
constituents = pd.read_excel(file_name, sheet_name = sheet)
constituents.head(10) # shows headers with top 5 rows
```

```python
joined_stocks = pd.merge(analysis_stocks, constituents, on ='Ticker', how ='left').copy()
name_col = joined_stocks.pop('Name')
joined_stocks.insert(0, name_col.name, name_col)  # Is in-place
joined_stocks.head(25)
```

```python
joined_stocks.columns
```

```python
from datetime import datetime
import openpyxl
import os

def save_excel_sheet(df, filepath, sheetname, index=False):
    # Create file if it does not exist
    if not os.path.exists(filepath):
        df.to_excel(filepath, sheet_name=sheetname, index=index)

    # Otherwise, add a sheet. Overwrite if there exists one with the same name.
    else:
        with pd.ExcelWriter(filepath, engine='openpyxl', if_sheet_exists='replace', mode='a') as writer:
            df.to_excel(writer, sheet_name=sheetname, index=index)

sheetname = datetime.now().strftime("%m-%d-%Y--%H-%M")

save_excel_sheet(joined_stocks, 'stock-data.xlsx', sheetname)
```
